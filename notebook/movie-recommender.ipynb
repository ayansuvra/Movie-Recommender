# Movie Recommender System

```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('movies.csv')
credits = pd.read_csv('credits.csv')
```

```python
movies = movies.merge(credits, on='title')
```

```python
movies.info()
```

```python
# choose important columns
# genres, movie_id, title, keywords, overview, cast, crew
```

```python
movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
movies.head()
```

```python
movies.info()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
type(movies['genres'][0])
```

```python
import ast
```

```python
def convert(obj):
    l = []
    obj = ast.literal_eval(obj)
    for i in obj:
        l.append(i['name'])
    return l
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['genres'].sample(5)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies['keywords'].sample(5)
```

```python
def convertcast(obj):
    l = []
    obj = ast.literal_eval(obj)
    count = 0
    for i in obj:
        if(count == 5):
            break
        l.append(i['name'])
        count += 1
    return l
```

```python
movies['cast'] = movies['cast'].apply(convertcast)
```

```python
def find_director(obj):
    l = []
    obj = ast.literal_eval(obj)
    for i in obj:
        if i['job'] == "Director":
            l.append(i['name'])
            break
    return l
```

```python
movies['crew'] = movies['crew'].apply(find_director)
```

```python
movies.head()
```

```python
movies.info()
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
# removing spaces
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
df = movies[['movie_id', 'title', 'tags']]
```

```python
df.head()
```

```python
# string -> lower
df['tags'] = df['tags'].apply(lambda x: " ".join(x))
```

```python
df['tags'] = df['tags'].apply(lambda x: x.lower())
```

```python
df['tags'][0]
```

```python
#stemming
from nltk.stem.porter import PorterStemmer
PS = PorterStemmer()
```

```python
def stemming(text):
    y = []
    for i in text.split():
        y.append(PS.stem(i))
    return " ".join(y)
```

```python
df['tags'].apply(stemming)
```

```python
# vectorization
from sklearn.feature_extraction.text import CountVectorizer
CV = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = CV.fit_transform(df['tags'])
```

```python
vectors = vectors.toarray()
```

```python
CV.get_feature_names_out().shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
df.shape, similarity.shape
```

```python
def recommend(movie):
    index = df[df['title'] == movie].index[0]
    pairs = list(enumerate(similarity[index].tolist()))
    pairs = sorted(pairs, key=lambda z:z[1], reverse=True)[1:11]
    for i in pairs:
        print(df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
